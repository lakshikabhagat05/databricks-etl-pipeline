```sql
MERGE INTO workspace.default.employee_silver tgt
USING workspace.default.employee_large_dataset  src
ON
    tgt.education = src.Education
AND tgt.joining_year = src.JoiningYear
AND tgt.city = src.City
AND tgt.gender = src.Gender

WHEN MATCHED THEN UPDATE SET
    tgt.payment_tier = src.PaymentTier,
    tgt.age = CAST(src.Age AS INT),
    tgt.ever_benched = src.EverBenched,
    tgt.experience_years = CAST(src.ExperienceInCurrentDomain AS INT),
    tgt.leave_flag = CAST(src.LeaveOrNot AS INT),
    tgt.total_company_experience = (YEAR(CURRENT_DATE) - src.JoiningYear),
    tgt.seniority_level =
        CASE 
            WHEN CAST(src.Age AS INT) < 25 THEN 'Junior'
            WHEN CAST(src.Age AS INT) BETWEEN 25 AND 35 THEN 'Mid'
            ELSE 'Senior'
        END

WHEN NOT MATCHED THEN
INSERT (
    education,
    joining_year,
    city,
    payment_tier,
    age,
    gender,
    ever_benched,
    experience_years,
    leave_flag,
    total_company_experience,
    seniority_level
)
VALUES (
    src.Education,
    src.JoiningYear,
    src.City,
    src.PaymentTier,
    CAST(src.Age AS INT),
    src.Gender,
    src.EverBenched,
    CAST(src.ExperienceInCurrentDomain AS INT),
    CAST(src.LeaveOrNot AS INT),
    (YEAR(CURRENT_DATE) - src.JoiningYear),
    CASE 
        WHEN CAST(src.Age AS INT) < 25 THEN 'Junior'
        WHEN CAST(src.Age AS INT) BETWEEN 25 AND 35 THEN 'Mid'
        ELSE 'Senior'
    END
);
```
