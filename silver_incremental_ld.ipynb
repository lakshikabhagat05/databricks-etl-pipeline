```sql
MERGE INTO workspace.default.employee_silver tgt
USING workspace.default.employee_raw src
ON tgt.emp_id = src.emp_id

WHEN MATCHED THEN UPDATE SET
    tgt.name = src.name,
    tgt.department = src.department,
    tgt.role = src.role,
    tgt.location = src.location,
    tgt.ctc = src.ctc,
    tgt.ctc_lakhs = ROUND(src.ctc / 100000, 2),
    tgt.join_date = src.join_date,
    tgt.experience_years = ROUND(DATEDIFF(CURRENT_DATE, src.join_date) / 365, 2)

WHEN NOT MATCHED THEN INSERT (
    emp_id,
    name,
    department,
    role,
    location,
    ctc,
    ctc_lakhs,
    join_date,
    experience_years
)
VALUES (
    src.emp_id,
    src.name,
    src.department,
    src.role,
    src.location,
    src.ctc,
    ROUND(src.ctc / 100000, 2),
    src.join_date,
    ROUND(DATEDIFF(CURRENT_DATE, src.join_date) / 365, 2)
);
```
