```sql
-- 1. Row count check
SELECT COUNT(*) AS total_records
FROM workspace.default.employee_silver;

-- 2. Null checks
SELECT
    SUM(CASE WHEN education IS NULL THEN 1 ELSE 0 END) AS null_education,
    SUM(CASE WHEN joining_year IS NULL THEN 1 ELSE 0 END) AS null_joining_year,
    SUM(CASE WHEN city IS NULL THEN 1 ELSE 0 END) AS null_city,
    SUM(CASE WHEN age IS NULL THEN 1 ELSE 0 END) AS null_age
FROM workspace.default.employee_silver;

-- 3. Age validity
SELECT *
FROM workspace.default.employee_silver
WHERE age < 18 OR age > 65;

-- 4. Experience sanity
SELECT *
FROM workspace.default.employee_silver
WHERE experience_years > total_company_experience;

-- 5. Domain value checks
SELECT DISTINCT gender
FROM workspace.default.employee_silver;

SELECT DISTINCT ever_benched
FROM workspace.default.employee_silver;

-- 6. Attrition flag check
SELECT *
FROM workspace.default.employee_silver
WHERE leave_flag NOT IN (0,1);

-- 7. Duplicate detection
SELECT
    education,
    joining_year,
    city,
    gender,
    COUNT(*) AS cnt
FROM workspace.default.employee_silver
GROUP BY education, joining_year, city, gender
HAVING COUNT(*) > 1;
```
