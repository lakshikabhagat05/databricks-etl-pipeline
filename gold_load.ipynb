```sql
CREATE OR REPLACE TABLE workspace.default.employee_gold AS
SELECT
    department,
    COUNT(*) AS total_employees,
    ROUND(AVG(ctc_lakhs), 2) AS avg_ctc_lakhs,
    ROUND(MAX(ctc_lakhs), 2) AS max_ctc_lakhs,
    ROUND(MIN(ctc_lakhs), 2) AS min_ctc_lakhs
FROM workspace.default.employee_silver
GROUP BY department;
select * from workspace.default.employee_gold;
```
