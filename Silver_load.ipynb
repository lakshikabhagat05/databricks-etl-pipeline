```sql
CREATE OR REPLACE TABLE workspace.default.employee_silver AS
SELECT
    emp_id,
    name,
    department,
    role,
    location,
    CAST(ctc AS BIGINT) AS ctc,
    ROUND(ctc / 100000, 2) AS ctc_lakhs,
    join_date,
    ROUND(DATEDIFF(CURRENT_DATE, join_date) / 365, 2) AS experience_years
FROM workspace.default.employee_raw
WHERE ctc IS NOT NULL;
select * from workspace.default.employee_silver;
```
