```sql
CREATE OR REPLACE TABLE workspace.default.employee_silver AS
SELECT
Education as education,
    JoiningYear as joining_year,
    City as city,
    PaymentTier as payment_tier,
    cast(Age as INT) as age,
    Gender as gender,
    EverBenched as ever_benched,
    cast(ExperienceInCurrentDomain as INT) as experience_years ,
    cast(LeaveOrNot as INT) as leave_flag,
    -- derived columns
    (year(current_date) - joining_year) AS total_company_experience,
    case
        when age < 25 then 'Junior'
        when age between 25 and 35 then 'Mid'
        else 'Senior'
    end as seniority_level
FROM workspace.default.employee_large_dataset where age is not null;
select * from workspace.default.employee_silver;
```
